```python
import pandas as pd

file_path = "unemployed.csv"
df = pd.read_csv(file_path)  # Replace with the correct encoding
```

```python
df = df.transpose()
df.columns = df.iloc[0]  # Set the first row as column names
df = df[1:]  # Drop the first row from the data
df
```

```python
import numpy as np
df = df.replace(',', '', regex=True)

# Step 2: Convert the DataFrame to float
df = df.astype(float)
```

```python
df.columns = df.columns.str.strip()
```

```python
df['fr_unem'] = df['Vacant']/df['Total:']
```
