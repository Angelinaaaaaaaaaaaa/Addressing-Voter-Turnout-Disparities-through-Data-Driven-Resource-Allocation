```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

# Initialize geolocator
geolocator = Nominatim(user_agent="geoapiExercises")

# Function to get lat, long for a county
def get_lat_lon(county_name):
    try:
        location = geolocator.geocode(f"{county_name} County, Georgia, USA")
        if location:
            return pd.Series([location.latitude, location.longitude])
        else:
            return pd.Series([None, None])
    except GeocoderTimedOut:
        return pd.Series([None, None])

# Apply the function to the County column
df[['latitude', 'longitude']] = df['County'].apply(get_lat_lon)

# Display the DataFrame
print(df)
df
```

```python

# Optionally save to a CSV file
df.to_csv('counties_with_lat_lon.csv', index=False)
```
